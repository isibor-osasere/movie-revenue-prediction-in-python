### Observation
outliers in: 
- Time_taken
- Twitter_hastags
- Avg_age_Actors

```python
def p_outliers (df, col):
    upper_limit = np.percentile (df[col], [99]) * 3
    lower_limit = np.percentile (df[col], [1]) * 0.3
    
    return f"upper_limit: {upper_limit}  ---  lower_limit: {lower_limit}"
```

```python
p_outliers (df, "Time_taken")
```

```python
df.Time_taken[(df.Time_taken > 637.2) | (df.Time_taken < 32)] = df.Time_taken.mean ()
```
